###  LeNet-5
Когда люди говорят, что глубокое обучение не является чем-то новым, и что концепции были описаны в 1980-х и 1990-х годах, LeNet-5 является одной из фундаментальных сетей, о которых они говорят. Мы используем keras API от Tensorflow для воссоздания LeNet-5 и классификации букв алфавита на американском языке жестов, за исключением "j" и " z " (которые полагаются на динамические жесты).

<img src="https://user-images.githubusercontent.com/54672403/84397611-81c2d780-ac07-11ea-8fe5-f762c536a025.png">

```python
import pandas as pd
import numpy as np 
import os

import tensorflow as tf
from tensorflow.keras.layers import (Activation, Dropout, Flatten,
                                     Dense, SpatialDropout2D, Conv2D, MaxPooling2D)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import ReduceLROnPlateau

import matplotlib.pyplot as plt

def draw_hist(history):
    hist = history.history
    x_arr = np.arange(len(hist['loss'])) + 1

    fig = plt.figure(figsize=(12, 4))
    ax = fig.add_subplot(1, 2, 1)
    ax.plot(x_arr, hist['loss'], '-o', label='Train loss')
    ax.plot(x_arr, hist['val_loss'], '--<', label='Validation loss')
    ax.legend(fontsize=15)
    ax.set_xlabel('Epoch', size=15)
    ax.set_ylabel('Loss', size=15)

    ax = fig.add_subplot(1, 2, 2)
    ax.plot(x_arr, hist['accuracy'], '-o', label='Train acc.')
    ax.plot(x_arr, hist['val_accuracy'], '--<', label='Validation acc.')
    ax.legend(fontsize=15)
    ax.set_xlabel('Epoch', size=15)
    ax.set_ylabel('Accuracy', size=15)
    plt.show()

UPPER_case= 'ABCDEFGHIKLMNOPQRSTUVWXY'
dim_x = 28
dim_y = 28
batch_size=32
```

```python
## data/labels
x_train = pd.read_csv('MNIST/sign_mnist_train.csv')
x_train.head()
x_train.shape
y_train = np.array(x_train['label'])
x_train.drop('label', axis = 1, inplace = True)
x_train = np.array(x_train.values)

print("\tData shapes", x_train.shape, y_train.shape, "classes: ", len(np.unique(y_train)))
print('\n',*np.unique(y_train))

classes = len(np.unique(y_train))
x_train = x_train.reshape((-1, dim_x,dim_y,1))

## skip 'j'
y_train[y_train>8] = y_train[y_train>8] - 1

## label + one-hot
y = np.zeros((np.shape(y_train)[0],len(np.unique(y_train))))
for ii in range(len(y_train)):
    y[ii,y_train[ii]] = 1
y_train = y

print("\n\t New shapes", x_train.shape, y_train.shape)
print("\t Shape 1 example", x_train[1].shape, y_train[1].shape)
```

```python
no_validation = int(0.1 * (x_train.shape[0])) # 27455 ~~> 2745

x_val = x_train[0:no_validation,...]
y_val = y_train[0:no_validation,...]

x_train = x_train[no_validation:,...]
y_train = y_train[no_validation:,...]

print('Train/Val shapes', x_train.shape, y_train.shape,'\n\t\t', x_val.shape, y_val.shape)

## image generators with mild augmentation
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   rotation_range=10,
                                   width_shift_range=0.05,
                                   height_shift_range=0.05,
                                   shear_range=0.1,
                                   zoom_range=0.075)

train_generator = train_datagen.flow(x=x_train, y=y_train,
                                     batch_size=batch_size,
                                     shuffle=True)


test_datagen = ImageDataGenerator(rescale=1./255)

val_generator = test_datagen.flow(x=x_val, y=y_val,
                                  batch_size=batch_size,
                                  shuffle=True)
```

## Le-Net5 *v1*

<img src='https://user-images.githubusercontent.com/54672403/84562166-ffe3c300-ad5a-11ea-8cfb-f8d891a774eb.png' width='900'>

<img src='https://user-images.githubusercontent.com/54672403/84562242-97e1ac80-ad5b-11ea-88f0-b5111630851b.jpg' width='900'>


**Input :** Flattened 784px images (n, 28, 28, 1)
<br>**Output:** 0 - 24

**First Layer:** Convolutional Layer (CONV1):
>**Parameters:**
<br>&nbsp; Input (N) = 28 
<br>&nbsp; Padding (P) = 2
<br>&nbsp; Filter (F) = 5 x 5
<br>&nbsp; Stride (S) = 1  
**Conv Operation:** $((N + 2P - F) / S) + 1 = ((28 + 4 - 5) / 1) + 1 = 28 x 28$
<br>&nbsp; Мы будем применять 6 фильтров/ ядер, так что мы получим размерный выход 28 x 28 x 6

**Second Layer:** Average Pooling Layer (POOL1):
>**Parameters:**
<br>&nbsp; Input (N) = 28
<br>&nbsp; Filter (F) = 2 x 2
<br>&nbsp; Stride (S) = 2
<br>**AVG Pooling Operation:** $((N + 2P -F) / S) + 1 = ((28 - 2) / 2) + 1 = 14 x 14$
<br>&nbsp; В конце этого объединения мы получим размерный выход 14 x 14 x 6

**Third Layer:** Convolutional Layer (CONV2):
>**Parameters:** 
<br>&nbsp; Input (N) = 14
<br>&nbsp; Filter (F) = 5 x 5
<br>&nbsp; Stride (S) = 1
<br>**Conv Operation:** $((N + 2P - F) / S) + 1 = ((14 - 5) / 1) + 1 = 10 x 10;$
<br> Мы будем применять 16 фильтров / ядер, так что мы получим размерный выход 10 x 10 x 16

**Fourth Layer:** Average Pooling Layer (POOL2):
>**Parameters:** 
<br>&nbsp; Input (N) = 10
<br>&nbsp; Filter (F) = 2 x 2
<br>&nbsp; Stride (S) = 2
<br> **AVG Pooling Operation:** $((N + 2P -F) / S) + 1 = ((10 - 2) / 2) + 1 = 5 x 5$
<br>&nbsp; В конце этого объединения мы получим размерный выход размером 5 х 5 х 16

**Fifth Layer:** Fully Connected layer(FC1):
>**Parameters:**
<br>&nbsp; W: 400 * 120
<br>&nbsp; b: 120
Мы будем иметь размерный выход 120 x 1 

**Sixth Layer:** Fully Connected layer(FC2):
>**Parameters:** 
<br>&nbsp; W: 120 * 84
<br>&nbsp; b: 84
<br> У нас будет размерный выход 84 x 1

**Seventh Layer:** Output layer(Softmax):
>**Parameters:** 
<br>&nbsp; W: 84 * 24
<br>&nbsp; b: 24

```python
model = Sequential()

model.add(Conv2D(filters=6, kernel_size=(5,5), 
                 padding='same', activation='relu', input_shape=(dim_x,dim_y,1)))
model.add(MaxPooling2D(pool_size=2,strides=2))
model.add(Conv2D(filters=16, kernel_size=(5,5), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=2,strides=2))
model.add(Flatten())
model.add(Dense(120, activation='relu'))
model.add(Dense(84,  activation='relu'))
model.add(Dense(classes, activation='softmax'))

model.summary()
```

```python
adam = Adam(lr=5e-4)
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=adam)
```

```python
# Set a learning rate annealer
reduce_lr = ReduceLROnPlateau(
              monitor='val_accuracy', 
              patience=3, verbose=1, 
              factor=0.2, min_lr=1e-6)
```

```python
steps_per_epoch = int(len(y_train)/batch_size)
max_epochs = 64

history = \
model.fit_generator(
         generator=train_generator,
         steps_per_epoch=steps_per_epoch,
         validation_data=val_generator,
         validation_steps=50,
         epochs=max_epochs,
         callbacks=[reduce_lr],
         verbose=2)
```

```python
draw_hist(history)
```

## Le-Net5 *v2*

```python
# define Le-Net5
model = Sequential()

model.add(Conv2D(filters=6, kernel_size=(5,5), strides=1,
                 input_shape=(dim_x,dim_y,1),
                 activation=tf.nn.tanh))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(Conv2D(filters=16, kernel_size=(5,5), strides=1,
                 activation=tf.nn.tanh))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(Flatten())
model.add(Dense(120, activation=tf.nn.tanh))
model.add(Dense(84,  activation=tf.nn.tanh))
model.add(Dense(classes, activation=tf.nn.softmax))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=1e-4), metrics=['accuracy'])
```

```python
steps_per_epoch = int(len(y_train)/batch_size)
max_epochs = 64
history = model.fit_generator(
         generator=train_generator,
         steps_per_epoch=steps_per_epoch,
         validation_data=val_generator,
         validation_steps=50,
         epochs=max_epochs,
         verbose=2)
```

```python
draw_hist(history)
```
