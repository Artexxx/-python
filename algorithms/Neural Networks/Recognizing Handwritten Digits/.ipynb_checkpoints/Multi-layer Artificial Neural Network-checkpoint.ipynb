```python
import os
import struct
import numpy as np
 
def load_df(path, kind='train'):
    labels_path = os.path.join(path, '%s_labels.idx1-ubyte' % kind)
    images_path = os.path.join(path, '%s_images.idx3-ubyte' % kind)
        
    with open(labels_path, 'rb') as lbpath:
        magic, n = struct.unpack('>II', lbpath.read(8))
        labels = np.fromfile(lbpath, dtype=np.uint8)

    with open(images_path, 'rb') as imgpath:
        magic, num, rows, cols = struct.unpack(">IIII", imgpath.read(16))
        images = np.fromfile(imgpath, dtype=np.uint8).reshape(len(labels), 784)
        images = ((images / 255.) - .5) * 2 # нормализация пикселей в диапазон [-1, 1]
 
    return images, labels
```

```python
X_train, y_train = load_df('image_data', kind='train')
print('TRAIN df: Rows: %d, columns: %d' % (X_train.shape))

X_test, y_test = load_df('image_data', kind='test')
print('TEST df:  Rows: %d, columns: %d' % (X_test.shape))
```

```python
fig, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True,)
ax = ax.flatten()
for i in range(10):
    img = X_train[y_train == i][0].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys')
ax[0].set_xticks([])
ax[0].set_yticks([]);
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5, sharex=True, sharey=True,)
ax = ax.flatten()
for i in range(25):
    img = X_train[y_train == 7][i].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys')

    ax[0].set_xticks([])
ax[0].set_yticks([]);
```

```python
from NeuralNetMLP import NeuralNetMLP 

nn = NeuralNetMLP(n_hidden=100, l2=0.01, 
                  max_iter=50, eta=0.0005,
                  minibatch_size=100, 
                  shuffle=True, random_state=1)

nn.fit(X_train=X_train[:55000], 
       y_train=y_train[:55000],
       X_valid=X_train[55000:],
       y_valid=y_train[55000:]);
```

```python
plt.plot(range(nn.max_iter), nn.eval_['cost'])
plt.ylabel('Ихдержки')
plt.xlabel('Эпохи');
```

```python
plt.plot(range(nn.max_iter), nn.eval_['train_acc'], 
         label='training')
plt.plot(range(nn.max_iter), nn.eval_['valid_acc'], 
         label='validation', linestyle='--')
plt.ylabel('Ихдержки')
plt.xlabel('Эпохи')
plt.legend();
```

```python
y_test_pred = nn.predict(X_test)
acc = (np.sum(y_test == y_test_pred)
       .astype(np.float) / X_test.shape[0])

print(f'Правильность предсказаний: {acc:.2%}')
```

```python
img = X_test[y_test != y_test_pred][:25]
correct_lab = y_test[y_test != y_test_pred][:25]
miscl_lab = y_test_pred[y_test != y_test_pred][:25]

fig, ax = plt.subplots(nrows=5, ncols=5, sharex=True, sharey=True,)
ax = ax.flatten()
for i in range(25):
    ax[i].imshow(img[i].reshape(28, 28), cmap='Greys', interpolation='nearest')
    ax[i].set_title(f'rl: {correct_lab[i]} pr: {miscl_lab[i]}')

ax[0].set_xticks([])
ax[0].set_yticks([]);
```
