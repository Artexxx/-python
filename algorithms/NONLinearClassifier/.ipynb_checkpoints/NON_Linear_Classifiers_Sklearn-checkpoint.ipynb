```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from plot_decision_regions import plot_decision_regions
```

# Подготовка данных для обучения 

```python
df = datasets.load_iris()
X_iris, y_iris = df.data[:, [2, 3]], df.target
```

```python
X_circl, y_circl = make_circles(n_samples=600, factor = 0.1, noise=0.3, random_state=0)
```

```python
np.random.seed(1)
X_xor = np.random.randn(200, 2)
y_raw = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_raw, 1, -1)
```

# Метод опорных векторов
`gamma` - параметр отсечения для гаусовской сферы
`ядро` - Radial Basis Function - гаусовое ядро

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf', random_state=42, gamma=0.1, C=10)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm), plt.show()

svm = SVC(kernel='rbf', random_state=42, gamma=0.1, C=10)
svm.fit(X_circl, y_circl)
plot_decision_regions(X_circl, y_circl, classifier=svm), plt.show();
```

```python
svm = SVC(kernel='rbf', random_state=42, gamma=0.1, C=10)
svm.fit(X_iris, y_iris)
plot_decision_regions(X_iris, y_iris, classifier=svm), plt.show();

svm = SVC(kernel='rbf', random_state=42, gamma=20, C=10)
svm.fit(X_iris, y_iris)
plot_decision_regions(X_iris, y_iris, classifier=svm), plt.show();
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=42)
tree.fit(X_iris, y_iris)
plot_decision_regions(X_iris, y_iris, classifier=tree)
```

# Random forest
- `n_estimators` - количество деревьев
- `n_jobs` -  количество ядер

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='gini', n_estimators=25, n_jobs=2, random_state=42)
forest.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=forest)
```

# K Nearest Neighbors 
простой ленивый алгоритм (отсутствуют издержки)<br>
`цель` - основываясь на расстояние найти наибольшее число соседей
- `p` регулирует расстояние<br>1 - расстояние Манхэттена, 2 - Минковского *(эвклидово)* ....

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5 ,metric ='minkowski',p=2)
knn.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=knn), plt.show();

knn.fit(X_circl, y_circl)
plot_decision_regions(X_circl, y_circl, classifier=knn), plt.show();
```
