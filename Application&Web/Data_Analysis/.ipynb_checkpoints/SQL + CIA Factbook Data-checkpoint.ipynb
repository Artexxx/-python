```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline

import warnings # Histogram error
warnings.filterwarnings("ignore")

import seaborn as sns
cm = sns.light_palette("pink", as_cmap=True)
def read_sql_query(q, conn):
    return pd.read_sql_query(q, conn)\
    .style.set_properties(**{'background-color': '#FFE7E7'})\
    .background_gradient(cmap=cm).hide_index()

conn = sqlite3.connect('factbook.db')
cursor = conn.cursor()
```

# Analyzing CIA Factbook Data

Демографические данные, собранные ЦРУ относительно всех стран мира, хранятся в формате базы данных. Информация из базы данных запрашивается с помощью SQLite для выполнения исследовательского анализа данных (EDA) и создания визуализаций.

В этом проекте я буду работать с [CIA World Factbook](https://www.cia.gov/library/publications/the-world-factbook/) -это данные, содержащие демографическую информацию о различных странах мира. Данные хранятся в базе данных, которую можно клонировать из [публичного репозитория] (https://github.com/factbook/factbook.sql/releases). SQL будет использоваться для взаимодействия с базой данных, чтобы сделать запрос данных, имеющих отношение к различным вопросам, представляющим интерес.

```python
# информация о таблицах в базе данных
q = "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(q, conn)
```

```python
cursor.execute(q).fetchall()
```

```python
# первые 5 строк из таблицы facts в базе данных
q2 = "SELECT * FROM facts LIMIT 5"
read_sql_query(q2, conn)
```

```python
# min population, max population, minimum population growth, maximum population growth
q3 = '''
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
'''
pd.read_sql_query(q3, conn)
```

Давайте рассмотрим более внимательно страны с населением 0 и 7256490011, поскольку эти значения кажутся нереалистичными.

```python
q4 = '''
SELECT *
FROM facts
WHERE population = 0
'''
pd.read_sql_query(q4, conn)
```

```python
q5 = '''
SELECT *
FROM facts
WHERE population = 7256490011
'''
pd.read_sql_query(q5, conn)
```

Похоже, что население Антарктиды составляет 0 человек.
Население строки "World" составляет 7256490011 человек, что, по-видимому, соответствует действительности.

```python
# запрос, исключающий две страны
q6 = '''
SELECT population, population_growth, birth_rate, death_rate
FROM facts
WHERE name <> "Antarctica" 
AND   name <> "World";
'''
hist_df = pd.read_sql_query(q6, conn)

# Histogram
style.use('ggplot')
hist_df.hist(figsize=(10,10))
plt.show()
```

```python
# Какие страны имеют самую высокую плотность населения
q7 = '''
SELECT name, CAST(population AS FLOAT) / CAST(area AS FLOAT) AS pop_density
FROM facts
ORDER BY pop_density DESC
LIMIT 10
'''
read_sql_query(q7, conn)
```

```python
pd.read_sql_query(q7, conn).hist()
plt.show()
```

```python
# Какие страны имеют самое высокое соотношение воды и суши
q8 = '''
SELECT name, CAST(area_water AS FLOAT) / CAST(area_land AS FLOAT) AS water_land_ratio
FROM facts
ORDER BY water_land_ratio DESC
LIMIT 10
'''
read_sql_query(q8, conn)
```

```python
# В каких странах воды больше, чем земли?
q8 = '''
SELECT name, CAST(area_water AS FLOAT) / CAST(area_land AS FLOAT) AS water_land_ratio
FROM facts
WHERE water_land_ratio > 1.0
ORDER BY water_land_ratio DESC
'''
read_sql_query(q8, conn)
```
