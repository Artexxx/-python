# Digital Media Store Sales Analysis

Реляционная база данных, представляющая собой магазин цифровых медиа, в США, запрашивается на основе важных деловых вопросов, касающихся продаж, персонала и маркетинга. Аналитические данные и визуализации извлекаются из запрашиваемых данных, что может позволить компании принимать более обоснованные бизнес-решения.

Этот проект сосредоточен на анализе реляционной базы данных, хранящей информацию о продуктах, клиентах, сотрудниках и транзакциях магазина цифровых медиа. Соответствующая информация будет запрошена из базы данных с помощью SQL и визуализирована для того, чтобы ответить на следующие вопросы:
- Какие альбомы следует приобрести в будущем, чтобы сохранить или улучшить продажи?
- Какие сотрудники генерируют больше всего продаж? Как это соотносится с другими сотрудниками?
- Какие страны имеют наибольший потенциал для роста продаж?
- Больше клиентов покупающих целые альбомы или клиентов приобретающих один или несколько отдельных треков?

![Database Schema](https://s3.amazonaws.com/dq-content/191/chinook-schema.svg)

```python
#~~~~~~~~~| HELPER FUNCTIONS |~~~~~~~~~~
import seaborn as sns
cm = sns.light_palette("pink", as_cmap=True)
def read_sql_query(q):
    with sqlite3.connect('data/chinook.db') as conn:
        return pd.read_sql_query(q, conn).round(3)\
        .style.set_properties(**{'background-color': '#FFE7E7'})\
        .background_gradient(cmap=cm)

# Запускает sql-запрос и возвращает dataframe
def run_query(q):
    with sqlite3.connect('data/chinook.db') as conn:
        return pd.read_sql(q, conn)

# Executes - sql command
def run_command(c):
    with sqlite3.connect('data/chinook.db') as conn:
        conn.isolation_level = None
        conn.execute(c)
        
# return: a list of tables & views
def show_tables():
    st = '''
    SELECT
        name,
        type
    FROM sqlite_master
    WHERE type IN ("table","view");
    '''
    return read_sql_query(st)
```

```python
show_tables()
```

- **(INNER) JOIN:** Returns records that have matching values in both tables
- **LEFT (OUTER) JOIN:** Return all records from the left table, and the matched records from the right table
- **RIGHT (OUTER) JOIN:** Return all records from the right table, and the matched records from the left table
- **FULL (OUTER) JOIN:** Return all records when there is a match in either left or right table

## Выбор новых альбомов для заказа
Чтобы выбрать новые альбомы, которые магазин должен приобрести в будущем, будет определено количество продаж, связанных с каждым жанром. Это позволит магазину принимать обоснованные решения относительно того, какие типы альбомов следует добавить, чтобы сохранить или улучшить продажи, сохраняя при этом гибкость в выборе.

```python
# return: каждый жанр с количеством пластинок и процентом от проданного 
numTracksByGenre = '''
WITH usa_tracks_sold AS 
    (
    SELECT
        g.name AS genre,
        il.quantity AS quantity
    FROM track AS t
    INNER JOIN genre g ON g.genre_id = t.genre_id
    INNER JOIN invoice_line il ON il.track_id = t.track_id
    INNER JOIN invoice  i ON i.invoice_id  = il.invoice_id
    INNER JOIN customer c ON c.customer_id = i.customer_id
    WHERE c.country = 'USA' 
    )

SELECT
    uts.genre AS genre,
    SUM(uts.quantity) AS tracks_sold,
    CAST(SUM(uts.quantity) AS FLOAT) / (SELECT COUNT(*) FROM usa_tracks_sold)*100 AS percentage_sold
FROM usa_tracks_sold AS uts
GROUP BY genre
ORDER BY tracks_sold DESC;
'''

read_sql_query(numTracksByGenre)
```

```python
tracksByGenre = run_query(numTracksByGenre)
tracksByGenre.set_index("genre", inplace=True, drop=True)
del tracksByGenre.index.name

style.use('fivethirtyeight')
tracksByGenre['tracks_sold'].plot.barh(figsize=(10,6), fontsize=12, colormap='Pastel1')
plt.xlabel('Number of Sales', fontsize=13)
plt.title('Sales By Genre in USA', fontsize=15)
plt.gca().yaxis.grid(False) # del x-line
```

Исходя из количества продаж по жанрам, существует больше стимулов для выбора рок, панк, метал и R&B/Soul альбомов. Поскольку рок охватывает более 50% от общего объема продаж, приобретение большего количества альбомов у рок-исполнителей должно быть приоритетным.

## Анализ эффективности продаж сотрудников
Поскольку каждый клиент из магазина присвоен к агенту по продажам, надо анализировать покупки клиентов, принадлежащих каждому сотруднику. Это поможет определить, какие сотрудники работают хуже или лучше других.

```python
# Запрос общей суммы продаж в долларах и дату найма, назначенных каждому агенту отдела продаж.
emp_perf = '''
WITH cust_invoices AS (
    SELECT 
        c.customer_id,
        SUM(i.total) customer_total,
        c.support_rep_id
    FROM invoice i
    LEFT JOIN customer AS c ON i.customer_id = c.customer_id
    GROUP BY c.customer_id 
    ORDER BY c.customer_id
    )
    
SELECT
    ci.support_rep_id,
    e.first_name || ' ' || e.last_name AS employee_name,
    SUM(ci.customer_total) AS total_sales,
    e.hire_date
FROM cust_invoices AS ci
LEFT JOIN employee e ON e.employee_id = ci.support_rep_id
GROUP BY ci.support_rep_id
ORDER BY total_sales DESC
'''

read_sql_query(emp_perf)
```

```python
employeePerf = run_query(emp_perf)
employeePerf = employeePerf[['employee_name','total_sales']]
employeePerf.set_index('employee_name', inplace=True, drop=True)

employeePerf['total_sales'].plot.barh(figsize=(10,4), fontsize=12, colormap='Pastel1')
plt.xlabel('Total Sales ($)', fontsize=13)
plt.ylabel('')
plt.title('Sales Breakdown by Employee', fontsize=15)
ax = plt.gca()
ax.yaxis.grid(False)
plt.show()
```

Понятно, что Джейн показала лучшие результаты в плане продаж, за ней последовала Маргарет, затем Стив. Возможно, это можно объяснить количеством опыта работы, основанного на датах их найма.

## Анализ продаж по странам
Чтобы проанализировать продажи по странам, запросим общее количество клиентов, общее значение продаж, среднее значение продаж на одного клиента и среднее значение заказа для каждой страны.

Результаты будут отсортированы по общему объему продаж в порядке убывания. Страны, у которых есть только один клиент, будут объединены в "другую" группу, расположенную cнизу.

```python
countrySales = '''
WITH country_info AS (
    SELECT
        c.country,
        COUNT(DISTINCT i.customer_id) AS num_customers,
        SUM(i.total) AS total_sales,
        SUM(i.total) / COUNT(DISTINCT i.customer_id) AS avg_sale_per_customer,
        AVG(i.total) AS avg_order
    FROM invoice i
    LEFT JOIN customer c ON c.customer_id = i.customer_id
    GROUP BY c.country
    ),

other_countries AS (
    SELECT 
        CASE
            WHEN ci.num_customers = 1 THEN 'Other'
            ELSE ci.country
        END AS country,
        ci.num_customers,
        ci.total_sales,
        ci.avg_sale_per_customer,
        ci.avg_order
    FROM country_info ci           
    )

SELECT
    country,
    SUM(num_customers) AS num_customers,
    SUM(total_sales) AS total_sales,
    AVG(avg_sale_per_customer) AS avg_sale_per_customer,
    AVG(avg_order) AS avg_order
FROM
    (
    SELECT
        oc.*,
        CASE
            WHEN oc.country = 'Other' THEN 1
            ELSE 0
        END AS sort
    FROM other_countries oc
    )
GROUP BY country
ORDER BY sort ASC, total_sales DESC;

'''

read_sql_query(countrySales)
```

```python
country_sales = run_query(countrySales)
country_sales.set_index('country' ,inplace=True)

# Sales by Country
fig, axes = plt.subplots(2, 2, figsize=(15, 13))
fig.set_facecolor('white')

# Вверху слева - доля от общего объема продаж
country_sales.plot.pie(
    y='total_sales',
    title='Total Sales Breakdown',
    ax=axes[0,0],
    legend = False,
    autopct='%1.1f%%',
    textprops={'fontsize': 10, 'color': 'k'}, 
    pctdistance=0.85, 
    labeldistance=1.1,
    cmap = 'Set3'
)
centre_circle = plt.Circle((0,0),0.40,fc='white')
axes[0,0].add_artist(centre_circle)
axes[0,0].axis('off')

# Top Right - customer % and sale %
cust_sale_pct = country_sales[['num_customers', 'total_sales']].copy()
cust_sale_pct['num_customers'] = cust_sale_pct['num_customers'] / cust_sale_pct['num_customers'].sum() * 100
cust_sale_pct['total_sales'] = cust_sale_pct['total_sales'] / cust_sale_pct['total_sales'].sum() * 100
cust_sale_pct.rename(columns={'num_customers': '% Customers', 'total_sales': '% Sales'}, inplace=True)
cust_sale_pct.index.rename('', inplace=True)
cust_sale_pct.plot.barh(
    ax=axes[0,1],
    title='Customers vs Sales\n Percentage Breakdown'
)
axes[0,1].yaxis.grid(False)
axes[0,1].set_xticklabels(labels = ['0', '10', '20', '30', '40', '50%'])
axes[0,1].yaxis.set_tick_params(labelsize=12)
axes[0,1].xaxis.set_tick_params(labelsize=12)
axes[0,1].legend(prop={"size":12})

# Bottom Left - Average Order (Percent Difference from mean)
avg_orders = country_sales[['avg_order']].copy()
mean_avg_order = avg_orders['avg_order'].mean()
avg_orders['order_pct_change'] = (avg_orders['avg_order'] / mean_avg_order - 1) * 100
avg_orders.index.rename('', inplace=True)
colors = [plt.cm.Set3(i) for i in np.linspace(0, 1, country_sales.shape[0])]
avg_orders.plot.barh(
    ax=axes[1,0],
    y='order_pct_change',
    title='Average Order\n Percent Change from Mean',
    legend=False,
    color=colors
)
axes[1,0].yaxis.grid(False)
axes[1,0].yaxis.set_tick_params(labelsize=12)
axes[1,0].xaxis.set_tick_params(labelsize=12)

# Внизу справа - средняя продажа($) на одного клиента
country_sales.plot.barh(
    ax=axes[1,1],
    y='avg_sale_per_customer',
    title='Average Sales per Customer\n (Dollars, $)',
    legend=False,
    color=colors
)
axes[1,1].yaxis.grid(False)
axes[1,1].set_ylabel('')
axes[1,1].yaxis.set_tick_params(labelsize=12)
axes[1,1].xaxis.set_tick_params(labelsize=12)
fig.tight_layout() # чтоб таблицы не пересекались
plt.show()
```

Из представленных показателей продаж по странам видно, что наибольший процент продаж и клиентов приходится на США, что вполне логично, поскольку медиамагазин в настоящее время базируется в США.

За пределами США - Чехия является хорошим кандидатом на проведение новых рекламных кампаний, так как страна генерирует больше всего денег на одного клиента. Другими потенциальными кандидатами на маркетинг за рубежом являются Индия, а также Великобритания.

## Альбомы vs отдельные треки
Далее мы определим пропорцию клиентских транзакций, которые составляют покупку целых альбомов, по сравнению с теми, которые состоят из коллекций отдельных треков.

```python
alb_trks = '''
WITH invoice_ref_track AS
    (
    SELECT
        il.invoice_id,
        MIN(il.track_id) ref_track
    FROM invoice_line il
    GROUP BY il.invoice_id
    )

SELECT
    is_album,
    COUNT(invoice_id) AS num_invoices,
    CAST(COUNT(invoice_id) AS FLOAT) / (SELECT COUNT(*) FROM invoice_ref_track) AS percent_invoices
FROM(
    SELECT
        irt.invoice_id,
        irt.ref_track,
        CASE
            WHEN
                (
                SELECT t.track_id
                FROM track t
                WHERE t.album_id = (
                                    SELECT t2.album_id
                                    FROM track t2
                                    WHERE t2.track_id = irt.ref_track
                                    )

                EXCEPT

                SELECT il.track_id
                FROM invoice_line il
                WHERE il.invoice_id = irt.invoice_id
                )IS NULL
            AND
                (
                SELECT il.track_id
                FROM invoice_line il
                WHERE il.invoice_id = irt.invoice_id

                EXCEPT

                SELECT t.track_id
                FROM track t
                WHERE t.album_id = (
                                    SELECT t2.album_id
                                    FROM track t2
                                    WHERE t2.track_id = irt.ref_track
                                    )

                )IS NULL
            THEN "yes"
            ELSE "no"       
        END AS is_album
    FROM invoice_ref_track AS irt)
GROUP BY is_album;
'''

read_sql_query(alb_trks)
```

Исходя из приведенного выше запроса, очевидно, что большая часть продаж приходится на то, что клиенты покупают коллекцию отдельных треков (81%), а не целый альбом.
