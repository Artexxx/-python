#### 2. Целые числа

$\mathbb{\mathbb{Z}}$ — множество целых чисел, $\mathrm{\mathbb{\mathbb{Z}}}=\left\{ \dots,-4,-3,-2,-1,0,1,2,3,4,\dots\right\} $.
<br>
&nbsp;&nbsp;Целые числа мы умеем складывать, вычитать и умножать.
<br>
&nbsp;&nbsp;Что значит $3+4$? Это значит: на числовой прямой обозначить тройку и приставить к ней отрезок, обозначающий 4. $3+4=7$. Также можно отложить 4 влево. Это будет $3-4=3+\left(-4\right)$. Содержательно это означает, что мы уменьшаем тройку на 4. $3-4=1$. Операция вычитания обратна операции сложения: $\left(3-4\right)+4=3$. Если температура сначала упала на 4 градуса, а затем поднялась на 4 градуса, то ничего не изменилось.
<br>
&nbsp;&nbsp;Это всё очевидно, но требует специального названия. Называется это словом «группа». Что такое группа мы рассмотрим на примере целых чисел, исходя из идеи, что все математические идеи нужно вводить сразу же.

#### 3. Что такое группа

Итак, что такое группа? И всё время помните, что вы знаете только целые числа. Целые числа образуют группу по следующим причинам. А потом мы абстрагируемся от целых чисел.
<br>
&nbsp;&nbsp;Целые числа — это некоторое множество, любые элементы которых можно складывать друг с другом $\left(\mathbb{Z},+\right)$. Сложение обладает несколькими свойствами: $a+b=b+a$. Но, как ни странно, это свойство в определение группы не входит, а группы, которые обладают этим свойством носят специальное название — коммутативные или Абелевы. Это равенство следует из геометрических соображений
<br>
&nbsp;&nbsp;Второе свойство: $\left(a+b\right)+c=a+\left(b+c\right)$. Это наглядно очевидно для целых чисел.
<br>
&nbsp;&nbsp;Есть некоторое специальное число 0, которое обладает следующим свойством $a+0=0+a=a$. Ноль при сложении ничего не меняет. Это называется нейтральный элемент группы.
<br>
&nbsp;&nbsp;Для любого целого $a$ существует противоположное ему $-a$, то есть такое, которое его нейтрализует$ a+\left(-a\right)=\left(-a\right)+a=0$.
<br>
&nbsp;&nbsp;Если мы возьмем эти три аксиомы (свойства) и будем их изучать их безотносительно к тому, что такое целые числа и сложение, а просто есть какое-то множество и какая-то операция. Неважно, конечное или бесконечное. Множество $G$, с элементами можно проделывать операцию $*$: чисто абстрактное действие, которое выполнимо с любыми элементами $G$. Заменив в описанных выше свойствах плюс на звездочку, мы можем изучать такую абстрактную группу. Многие выводы можно сделать для любых таких множеств, независимо от природы элементов, а не только для целых чисел.

#### 4 Целые числа как группа

Мы будем их изучать немного необычно: как группу. Потом мы соберём все плоды с этого дерева, сейчас мы его посадим. Мы сажаем такое дерево, у себя в голове, что целые числа — это группа, и задавать надо теоретико-групповые вопросы. Если в целых числах такие подмножества $\mathbb{\mathbb{Z}}\supset A$, которые то же образуют группу относительно той же операции сложения $\left(A,+\right)$, то есть подгруппы. Операция не выводит за рамки группы.
<br>
&emsp;&emsp;Одну такую подгруппу можно угадать. Это множество $\left\{ 0\right\}$. Есть ли другие подгруппы?
<br>
&emsp;&emsp;Предположим, что подгруппа содержит некоторое число $b\neq0$. Я сразу могу сказать, что если подгруппа содержит b, то она содержит и ноль: $b-b=0$. Любая подгруппа целых чисел содержит ноль. Подгруппа точно содержит $\left\{ -b,0,b\right\}$. Если мы можем вычесть из нуля b, то можем вычесть b и из результата. Это самый главный божий дар, который есть у людей — рекурсия.
<br>
&emsp;&emsp;Если мы рассматриваем выражение $b+b+b$, то мы можем вычислить его двумя способами: $\left(b+b\right)+b$ и $b+\left(b+b\right)$. Вдруг окажется, что это два разных числа? Но нет, они будут одинаковыми в силу аксиомы ассоциативности.
<br>
&emsp;&emsp;При $b=1$ получим все целые числа. Но можно взять b любое целое! Например, $b=13$. Получаем кратные 13. Мы приходим к операции умножения. Умножить $b$ на 3 это значит сложить b три раза: $3\cdot b=b+b+b$. Умножить $b$ на -4 — это взять 4 раза $-b$: $\left(-4\right)\cdot b=\left(-b\right)+\left(-b\right)+\left(-b\right)+\left(-b\right)$.
<br>
&emsp;&emsp;Рассматривая естественную конструкцию подгруппы в целых числах, мы пришли к понятию произведения целых чисел. Произведение — это следствие чисто групповой идеи. Группа является базовым понятием для всей современной математики. Поэтому с него и надо начинать.
<br>
&emsp;&emsp; $n\cdot b$ — это сумма $n$ слагаемых, каждое из которых равно $b$.
<br>
&emsp;&emsp;А если мы рассмотрим $b\cdot n$, то это $b$ слагаемых, каждое из которых равно $n$. Это одно и то же?

#### 5. Коммутативность умножения, подгруппы целых чисел

Итак, я сказал, что $n$ раз взятое число $b$ почему-то должно быть равно $b$ раз взятому числу $n$. Почему это так? А вот почему. Нарисуем прямоугольную таблицу. Она состоит из $n$ строк и $b$ столбцов. Просто потому, что я хочу. Пока рисуем, я хочу сказать, что математик это самая свободная профессия. Вот я хочу нарисовать такой прямоугольник и ни одна собака не может менять спросить, почему я вообще так решил и имею право. Все математические доказательства проводятся так. И если человек постоянно спрашивает, а почему вот это рассмотрели, то он не математик и ему нужно себя лечить от этого.
<br>
&emsp;&emsp;Итак $n\cdot b$ — это количество ячеек во всех строках этой таблицы, то есть количество ячеек во всей этой таблице. А $b \cdot n$ — это количество ячеек во всех столбцах этой таблицы. Но это одно и то же! Отсюда и следует это фундаментальное тождество.
<br>
&emsp;&emsp;А мы переходим к вопросу о том, все ли подгруппы целых чисел мы описали.
<br>
&emsp;&emsp;Итак, рассмотрим все целые числа на числовой прямой, отметим ноль и единицу. И вот очевидный пример подгруппы, когда мы взяли числа через одно и то же расстояние. Или, иными словами, мы взяли все кратные какого-то числа $b$. В примере $b=3$. Вот такое разряженное в три раза множество целых чисел носит, по очевидным причинам, название $ 3\cdot\mathbb{Z}$. А в общем случае — $b \cdot\mathbb{Z}$. Есть ещё подгруппа, состоящая из одного нуля.
<br>
&emsp;&emsp;Утверждение. Никаких других подгрупп в $\mathbb{Z} $ нет.
<br>
&emsp;&emsp;Доказательство. При доказательстве мы придём к ещё одной важной конструкции — делении с остатком.
<br>
&emsp;&emsp;Рассмотрим любую подгруппу. Если в ней только ноль, то мы её уже описали. Если есть некоторое $b$, то есть все его кратные. По крайней мере, любая подгруппа должна содержать эти кратные. Противоположный элемент тоже есть в подгруппе. Таким образом, в нетривиальной подгруппе справа от нуля есть какие-то элементы. Так давайте возьмём ближайший элемент к нулю и обозначим его $c$. Я утверждаю, что $b$ кратно $c$: $b=nc$. Докажем это.
