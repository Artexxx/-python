#### 2. Целые числа

$\mathbb{\mathbb{Z}}$ — множество целых чисел, $\mathrm{\mathbb{\mathbb{Z}}}=\left\{ \dots,-4,-3,-2,-1,0,1,2,3,4,\dots\right\} $.
<br>
&nbsp;&nbsp;Целые числа мы умеем складывать, вычитать и умножать.
<br>
&nbsp;&nbsp;Что значит $3+4$? Это значит: на числовой прямой обозначить тройку и приставить к ней отрезок, обозначающий 4. $3+4=7$. Также можно отложить 4 влево. Это будет $3-4=3+\left(-4\right)$. Содержательно это означает, что мы уменьшаем тройку на 4. $3-4=1$. Операция вычитания обратна операции сложения: $\left(3-4\right)+4=3$. Если температура сначала упала на 4 градуса, а затем поднялась на 4 градуса, то ничего не изменилось.
<br>
&nbsp;&nbsp;Это всё очевидно, но требует специального названия. Называется это словом «группа». Что такое группа мы рассмотрим на примере целых чисел, исходя из идеи, что все математические идеи нужно вводить сразу же.

#### 3. Что такое группа

Итак, что такое группа? И всё время помните, что вы знаете только целые числа. Целые числа образуют группу по следующим причинам. А потом мы абстрагируемся от целых чисел.
<br>
&nbsp;&nbsp;Целые числа — это некоторое множество, любые элементы которых можно складывать друг с другом $\left(\mathbb{Z},+\right)$. Сложение обладает несколькими свойствами: $a+b=b+a$. Но, как ни странно, это свойство в определение группы не входит, а группы, которые обладают этим свойством носят специальное название — коммутативные или Абелевы. Это равенство следует из геометрических соображений
<br>
&nbsp;&nbsp;Второе свойство: $\left(a+b\right)+c=a+\left(b+c\right)$. Это наглядно очевидно для целых чисел.
<br>
&nbsp;&nbsp;Есть некоторое специальное число 0, которое обладает следующим свойством $a+0=0+a=a$. Ноль при сложении ничего не меняет. Это называется нейтральный элемент группы.
<br>
&nbsp;&nbsp;Для любого целого $a$ существует противоположное ему $-a$, то есть такое, которое его нейтрализует $a+\left(-a\right)=\left(-a\right)+a=0$.
<br>
&nbsp;&nbsp;Если мы возьмем эти три аксиомы (свойства) и будем их изучать их безотносительно к тому, что такое целые числа и сложение, а просто есть какое-то множество и какая-то операция. Неважно, конечное или бесконечное. Множество $G$, с элементами можно проделывать операцию $*$: чисто абстрактное действие, которое выполнимо с любыми элементами $G$. Заменив в описанных выше свойствах плюс на звездочку, мы можем изучать такую абстрактную группу. Многие выводы можно сделать для любых таких множеств, независимо от природы элементов, а не только для целых чисел.

#### 4 Целые числа как группа

Мы будем их изучать немного необычно: как группу. Потом мы соберём все плоды с этого дерева, сейчас мы его посадим. Мы сажаем такое дерево, у себя в голове, что целые числа — это группа, и задавать надо теоретико-групповые вопросы. Если в целых числах такие подмножества $\mathbb{\mathbb{Z}}\supset A$, которые то же образуют группу относительно той же операции сложения $\left(A,+\right)$, то есть подгруппы. Операция не выводит за рамки группы.
<br>
&emsp;&emsp;Одну такую подгруппу можно угадать. Это множество $\left\{ 0\right\}$. Есть ли другие подгруппы?
<br>
&emsp;&emsp;Предположим, что подгруппа содержит некоторое число $b\neq0$. Я сразу могу сказать, что если подгруппа содержит b, то она содержит и ноль: $b-b=0$. Любая подгруппа целых чисел содержит ноль. Подгруппа точно содержит $\left\{ -b,0,b\right\}$. Если мы можем вычесть из нуля $b$, то можем вычесть $b$ и из результата. Это самый главный божий дар, который есть у людей — рекурсия.
<br>
&emsp;&emsp;Если мы рассматриваем выражение $b+b+b$, то мы можем вычислить его двумя способами: $\left(b+b\right)+b$ и $b+\left(b+b\right)$. Вдруг окажется, что это два разных числа? Но нет, они будут одинаковыми в силу аксиомы ассоциативности.
<br>
&emsp;&emsp;При $b=1$ получим все целые числа. Но можно взять $b$ любое целое! Например, $b=13$. Получаем кратные 13. Мы приходим к операции умножения. Умножить $b$ на 3 это значит сложить $b$ три раза: $3\cdot b=b+b+b$. Умножить $b$ на -4 — это взять 4 раза $-b$: $\left(-4\right)\cdot b=\left(-b\right)+\left(-b\right)+\left(-b\right)+\left(-b\right)$.
<br>
&emsp;&emsp;Рассматривая естественную конструкцию подгруппы в целых числах, мы пришли к понятию произведения целых чисел. Произведение — это следствие чисто групповой идеи. Группа является базовым понятием для всей современной математики.
<br>
&emsp;&emsp; $n\cdot b$ — это сумма $n$ слагаемых, каждое из которых равно $b$.
<br>
&emsp;&emsp;А если мы рассмотрим $b\cdot n$, то это $b$ слагаемых, каждое из которых равно $n$. Это одно и то же?

#### 5. Коммутативность умножения, подгруппы целых чисел

Итак, я сказал, что $n$ раз взятое число $b$ почему-то должно быть равно $b$ раз взятому числу $n$. Почему это так? А вот почему. Нарисуем прямоугольную таблицу. Она состоит из $n$ строк и $b$ столбцов. Просто потому, что я хочу. Пока рисуем, я хочу сказать, что математик это самая свободная профессия. Вот я хочу нарисовать такой прямоугольник и ни одна собака не может менять спросить, почему я вообще так решил и имею право. Все математические доказательства проводятся так. И если человек постоянно спрашивает, а почему вот это рассмотрели, то он не математик и ему нужно себя лечить от этого.
<br>
&emsp;&emsp;Итак $n\cdot b$ — это количество ячеек во всех строках этой таблицы, то есть количество ячеек во всей этой таблице. А $b \cdot n$ — это количество ячеек во всех столбцах этой таблицы. Но это одно и то же! Отсюда и следует это фундаментальное тождество.
<br>
&emsp;&emsp;А мы переходим к вопросу о том, все ли подгруппы целых чисел мы описали.
<br>
&emsp;&emsp;Итак, рассмотрим все целые числа на числовой прямой, отметим ноль и единицу. И вот очевидный пример подгруппы, когда мы взяли числа через одно и то же расстояние. Или, иными словами, мы взяли все кратные какого-то числа $b$. В примере $b=3$. Вот такое разряженное в три раза множество целых чисел носит, по очевидным причинам, название $ 3\cdot\mathbb{Z}$. А в общем случае — $b \cdot\mathbb{Z}$. Есть ещё подгруппа, состоящая из одного нуля.
<br>
&emsp;&emsp;Утверждение. Никаких других подгрупп в $\mathbb{Z} $ нет.
<br>
&emsp;&emsp;Доказательство. При доказательстве мы придём к ещё одной важной конструкции — делении с остатком.
<br>
&emsp;&emsp;Рассмотрим любую подгруппу. Если в ней только ноль, то мы её уже описали. Если есть некоторое $b$, то есть все его кратные. По крайней мере, любая подгруппа должна содержать эти кратные. Противоположный элемент тоже есть в подгруппе. Таким образом, в нетривиальной подгруппе справа от нуля есть какие-то элементы. Так давайте возьмём ближайший элемент к нулю и обозначим его $c$. Я утверждаю, что $b$ кратно $c$: $b=nc$. Докажем это.

#### 6. Подгруппы целых чисел

Мы изучаем подгруппы внутри целых чисел, то есть подмножества, замкнутые относительно операций сложения и вычитания.
<br>
&emsp;&emsp;Мы уже столкнулись с тем, что если мы возьмем любое число $b$, то все кратные этого числа образуют подгруппу. И мы заметили, что эта группа подозрительно похожа на группу целых чисел. Позже мы скажем «изоморфна», а пока мы не будем произносить это страшное слово. Попробуем доказать, что любая подгруппа имеет такой вид.
<br>
&emsp;&emsp;Подгруппа — непустое множество. Если кроме нуля в этой подгруппе ничего нет, то мы получаем $\left\{ 0\right\}$. Если же есть что-то кроме нуля: $b\neq0$, то есть и противоположный элемент $-b$. Соответственно, если есть какой-то не нулевой элемент, то есть и положительный. Теперь мы возьмём самый маленький положительный элемент $c$. Докажем, что все остальные элементы являются его кратными. Пусть $b$ не кратен $c$, то есть $ b\neq nc$. Так как вместе с $b$ в подгруппу входит и $-b$, то мы можем сразу считать, что $b$ — положительный. Пусть $b$ лежит между $nc$ и $\left(n+1\right)c$. Тогда $b-nc$ тоже принадлежит подгруппе. Так как $nc<b<\left(n+1\right)c$, то $0<b-nc<c$, таким образом, мы нашли положительный элемент, меньший c, что противоречит определению этого элемента $c$. Значит, предположение, что $b$ не кратно c приводит к противоречию и не верно. Значит любая подгруппа целых чисел является одной из подгрупп вида $b\cdot\mathbb{Z}$.

#### 7. Деление с остатком

Важнейшая операция с целыми числами, унаследованная нами от древней Греции. Пусть имеются числа $c$ и $b>c$, мы искали кратные $c$, которые с двух сторон охватывают $b$. Если бы нам повезло, то $b$ могло бы просто быть кратным $c$, тогда бы мы сказали, что $b$ просто делится на $c$ и всё. Если нам не повезло, то $b$ не кратное $c$. Тогда есть аксиома (или просто визуально видно), что существует целое $n$ такое, что $nc\le b<\left(n+1\right)c$. Тогда $d=b-nc$ называется остатком от деления $b$ на $c$. Остаток, по соглашению, всегда положителен. Но должен предупредить, что лучше избавиться от этого отягощающего фантома из школьной программы. Остаток, может быть и больше и меньше нулю, но по модулю он меньше $c$: $\left|d\right|<c$. Когда вы избавитесь от того, что учат в школе, то окажется, что любое число делиться с остатком двумя способами $b=nc+d$ и $n=\left(n+1\right)c-\left(c-d\right)$. Оба способа, в общем-то равноправны с точки зрения теории делимости. Самое важное, чтобы остаток по модулю был бы меньше, чем то на что мы делим. Посмотрим, что мы узнали, что если есть два целых числа $b$ и $c\neq0$, то всегда мы можем разделить $b$ на $c$ нацело или с небольшим остатком. Этот остаток будет всегда меньше $c$ по модулю.

#### 8. На подступах к основной теореме арифметики (ОТА)

Возьмём любое целое число, кроме нуля, единицы и минус единицы. Например, 36. И давайте попробуем понять, кратно ли оно кому-нибудь до него. В принципе, можно, например: $36=12+12+12=3\cdot12$. А тройка кратна каким-нибудь из предыдущих чисел: двойке? Тогда мы называем 3 простым числом. Простое число — это такое, которое делится только на единицу и само на себя (это самые простые случаи). А 12? 12 кратно двум. $36=3\cdot12=3\cdot2\cdot6$. А дальше можно продолжить этот процесс? У двойки допустимых предыдущих нет. Остаётся 6, которая кратна 2 и 3. $36=3\cdot2\cdot2\cdot3$.
<br>
&emsp;&emsp;А в другой комнате сидит другой человек, который говорит $36=6\cdot6=2\cdot3\cdot2\cdot3$. Как видим, разложения состоят из одних и тех же множителей. Перемешанных по разному, но мы уже знаем, что от перестановки множителей произведение не меняется.
<br>
&emsp;&emsp;Ещё кто-то мог заметить, что $36=4\cdot9$ и он нашёл совершенно новое разложение. Но, на самом деле, ничего нового в нём нет, так как $4=2\cdot2$ и $9=3\cdot3$, поэтому $36=4\cdot9=2\cdot2\cdot3\cdot3$. Опять все то же самое. Но может быть нам пока так везёт. Может быть какой-нибудь зверь, типа 1728 или что-то в этом духе, получится разными способами. Экспериментально человечеству не удалось обнаружить, чтобы два разных набора простых чисел (кроме просто перестановки) давали один и тот же результат.
<br>
&emsp;&emsp;В детских учебниках пишут, что это вот такое свойство целых чисел и просто примите его как данность. Но мы ничего как данность принимать не будем. Мы докажем его как основную теорему арифметики.
<br>
&emsp;&emsp;Мы могли бы ещё написать $36=\left(-3\right)\cdot\left(-2\right)\cdot2\cdot3$. Мы могли бы сказать, что это то же, что и $36=3\cdot2\cdot2\cdot3$. Но это нужно сначала чётко сформулировать. Мы можем заметить, что 3 и -3 ведут себя одинаково относительно умножения. Все кратные 3 — это кратные и -3, и так далее. Нужно как-то математически точно сформулировать, что мы имеем в виду, когда говорим, что любое число существенно единственным образом раскладывается на произведение простых. Ну и конечно надо бы это доказать. И для доказательства нам придётся ещё довольно сильно попотеть.

#### 9. Формулировка основной теоремы арифметики

Давайте оттачивать наш логический аппарат и сформулируем её строже, чем всё, что делали до этого.
<br>
&emsp;&emsp;Пусть $N\in\mathbb{Z}\setminus\left\{ -1,0,1\right\}$ . Если $N=p_{1}\cdot p_{2}\cdot\ldots\cdot p_{n}=q_{1}\cdot q_{2}\cdot \ldots\cdot q_{m}$, где $p_{i}$ и $q_{i}$ — простые числа. То
<br>
&emsp;&emsp;1) $n=m$
<br>
&emsp;&emsp;2) Существует перестановка набора $p_{1},p_{2},\ldots,p_{n}$, такая, что после неё получается $p_{i}^{\prime}=\pm q_{i}$. Иными словами, количества сомножителей совпадают и для любого $i$ существует $j$ такой, что $p_{i}=\pm q_{j}$. Причём все $j$ разные.
<br>
&emsp;&emsp;Будем доказывать. Это теорема, на которой как на ките, стоит оставшаяся часть математики. Так что сейчас мы её будем строго, формально доказывать. Это потребует некоторого интеллектуального усилия.

#### 10. ОТА: лемма о кузнечике

Кирпичём для нашей теоремы служит лемма о кузнечиках. Пусть у кузнечика 2 ноги (лапы). Прыжок с левой ноги всегда на $a$ шагов (единиц измерения), а с правой ноги — всегда на $b$. Тогда множество всех возможных точек попадания равно $d\mathbb{Z}$ при некотором $d$.
<br>
&emsp;&emsp;Что эта лемма утверждает? Казалось бы, множество точек попадания — это какое-то сложное множество. Его можно записать как $\left\{ ma+nb|m,n\in\mathbb{Z}\right\}$. Оно дословно совпадает с $d\mathbb{Z}$. Иными словами ему можно отрезать ноги и приделать одну ногу, которая скачет на d. От этого его возможности скакать не изменяться.
<br>
&emsp;&emsp;Приём доказательства будет нам знаком. Именно с помощью него мы доказывали, что все подгруппы целых чисел имеют такой вид.

#### 11. Доказательство леммы о кузнечике

Для доказательства сделает утверждения. Множество точек попадания $\left\{ ma+nb|m,n\in\mathbb{Z}\right\}$  — подгруппа во множестве всех целых числа. Смотрите, вот это уже логика чистой и высокой математики на простейшем примере. Я утверждаю, что если есть какие-то две точки попадания, то их сумма — тоже точка попадания. Разность тоже. Просто меняем все скачки налево меняем на скачки направо и наоборот. А следовательно она имеет вид $d\mathbb{Z}$. Видите, как работает математика. Мы когда-то зачем-то доказывали какой вид имеют подгруппы целых чисел. А теперь взяли какое-то специальное подмножество и доказали, что оно тоже имеет такой вид. Но это доказательство для того, кто уже родился гениальным математиком, для которого все эти конструкции естественны. Предположим, что Вы не гениальный математик.
<br>
&emsp;&emsp;Рассмотрим самое левое место, в которое кузнечик может попасть, как $d$. То есть в ноль и $d$ он может попасть, а между ними залезть не может. Понятно, что такая точка существует. Так как он точно может попасть в $a$. Так как все точки попадания целые, то самая малая положительная среди них найдётся. Я утверждаю, что как $a$, так и $b$ должны быть кратны этому $d$. Доказательство состоит в том, что если бы $a$ не было бы кратно $d$, то вот этот коротенький отрезок был бы доступен нашему кузнечику. На самом деле, быть такого не могло, поэтому все точки попадания кратны $d$. Мы имеем два важных свойства: $a=dl$ и $b=dk$. Но есть дополнительно очень важно свойство, которое надо здесь выписать: и при этом $d$ как точка попадания, можно записать как $d=am+bn$. $d$ — является общим делителем $a$ и $b$, то если мы рассмотрим какой-то ещё делитель $a$ и $b$, то из этой формулы ясно, что $d$ должен на другой этот делитель делится. Если $d^{\prime}|a$ и $d^{\prime}|b$, то $d^{\prime}|d$, а следовательно $\left|d^{\prime}\right|\le\left|d\right|$. Поэтому d оказывается самым большим общим делителем.
<br>
&emsp;&emsp;Итак, мы ввели очень важно понятие. d называется наибольшим общим делителем чисел a и b, не равных одновременно нулю.
<br>
&emsp;&emsp;Если одно из этих чисел равно нулю, это ещё полбеды. А вот если оба... Кстати, сами подумайте, почему дозволяется, чтобы одно из чисел было равно нулю?
<br>
&emsp;&emsp;Существенный случай, это когда оба не равны нулю. Тогда у них есть наибольший общий делитель. Это утверждение будет ключевым при доказательстве ОТА.

#### 12, Определение группы

Пусть $a,b\in\mathbb{Z}$, не равные нулю одновременно, то есть неверно, что оба равны нулю. Тогда существует такое целое $d>0$, что выполнены два свойства:
<br>
&emsp;&emsp;1. $d|a$ и $d|b$
<br>
&emsp;&emsp;2. Существуют $n,m\in\mathbb{Z}$ такие, что $d=ma+nb$.
<br>
&emsp;&emsp;Из пункта 2 следует, что $d$ — наибольший общий делитель a и b.
<br>
&emsp;&emsp;Если $a=p$ — простое и $p\nmid b$, то существуют такие $n,m\in\mathbb{Z}$, что $mp+nb=1$. Это основной кирпичик для ОТА. Почему равно единице? Потому что какой может быть общий делитель у простого $p$ и $b$. У $p$ делители $\left\{ -p,-1,1,p\right\}$ , а у $b$ нет делителя $p$, поэтому у $p$ и $b$ нет никаких нетривиальных общих делителей.
<br>
&emsp;&emsp;Теперь мы можем перейти к доказательству ОТА, но я хочу злоупотребить вашим вниманием. Я хочу, сформулировать что такое группа чисто абстрактно. Чтобы вы поняли, что целые числа — действительно группа по сложению.
<br>
&emsp;&emsp;Пусть дано непустое множество $G$ вместе с операцией $*$ (Операция должна быть применимой для любой пары элементов $G$), называется группой, если
<br>
&emsp;&emsp;1. $\left(a*b\right)*c=a*\left(b*c\right)$ — ассоциативность
<br>
&emsp;&emsp;2. Существует нейтральный элемент $e$: $e*a=a*e=a$.
<br>
&emsp;&emsp;3. Для любого $a\in G$ существует $b$ (нейтрализатор) такой, что $a*b=b*a=e$, он называется $a^{-1}$.
<br>
&emsp;&emsp;Если ещё всегда $a*b=b*a$, то группа называется абелевой или коммутативной.
<br>
&emsp;&emsp;Все целые числа образуют абелеву группу по сложению.
<br>
&emsp;&emsp;Операция в абстрактной алгебре может называться хоть горшком. Если операция оказывается плюсом, то $a^{-1}$ обозначают $-a$.

#### 13. Доказательство ОТА

Последний шаг — это так называемая основная лемма.
<br>
&emsp;&emsp;Пусть $p$ — простое, тогда если $a$ не делится на $p$ и $b$ не делится на $p$, то $ab$ тоже не делится на $p$. Это можно распространить на несколько чисел: если несколько чисел не делятся на простое $p$, то и их произведение не делится на $p$.
<br>
&emsp;&emsp;Я переформулирую это утверждение в обратном виде. Утверждения $A\to B$ и $\bar{B}\to\bar{A}$ — это одно и то же. Мы потом проведём отдельный урок логических утверждений.
<br>
&emsp;&emsp;Пойдём от противного. Пусть есть такие $a$ и $b$, что $ab$ делится на $p$, но при этом ни $a$, ни $b$ не делится на $p$.
<br>
&emsp;&emsp;Пусть $ab$ делится на $p$ и $a$ не делится на $p$. Отсюда нужно вывести, что $b$ делится на $p$. Так как $a$ не делится на $p$, то $1=am+pn$ при некоторых целых $m,n$. $b=abm+pbn$. По условию $ab$ делится на $p$, то есть $ab=pk$, то есть $b=pkm+pbn=p\left(km+bn\right)$, то есть $b$ — делится на $p$.
<br>
&emsp;&emsp;Таким образом, если все числа некоторого набора не делятся на $p$, то и их произведение не делится на $p$.
<br>
&emsp;&emsp;
Переходим к доказательству ОТА
<br>
&emsp;&emsp;
Если ОТА не верна, то есть числа, которые разными способами представляется как произведение простых. Среди них есть самое маленькое, обозначим его $N$:
<br>
&emsp;&emsp;$N=p_{1}\cdot\dots\cdot p_{k}=q_{1}\cdot\dots\cdot q_{l}$,причём эти разложения не получаются друг из друга перестановкой, тогда $\left\{ p_{1}\cdot\dots\cdot p_{k}\right\} \cap\left\{ q_{1}\cdot\dots\cdot q_{l}\right\} =\emptyset$. Если есть пересечение, то на него можно сократить, и мы получили бы меньшее число с двумя разными разложениями.
<br>
&emsp;&emsp;Тогда $p_{1}$ не может делиться на $q_{1}$, и вообще все $p_{i}$ не делятся на $q_{1}$, но тогда и $N$ не делится на $q_{1}$ по основной лемме. Но это противоречит тому, что $N=q_{1}\cdot\left(q_{2}\cdot\dots\cdot q_{l}\right)$. Теорема доказана.
