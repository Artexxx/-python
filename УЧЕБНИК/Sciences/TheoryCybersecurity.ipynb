### Основные понятия

>**Безопасность**  - это сохранение свойств информации таких как целостность, доступность и конфиденциальность.

>**Уязвимость** (_valnerability_) - это конкретный недостаток или свойства конкретной системы, которая может обладать этим недостатком. 

>**Угроза** (_threat_) - это потенциальная опасность для информационной системы в случае эксплуатации этой самой уязвимости 
<br>Угрозы существуют независимо от уязвимостей, и потенциальная опасность присутствует всегда, когда вы пользуйтесь информационными системами.

>**Безопасность** или **защита** - это комплекс организационных и технических мер и рубежей защиты на различных уровнях. 
<br>Выбирая идеальную защиту, стоит выстроить ее в виде нескольких рубежей 

>**Цель защиты информации** - сделать так, чтобы стоимость атаки была дороже, чем выгода от реализации этой атаки. 

>**Threat Hunting** - активный поиск угроз в инфраструктуре.
<br>  Поиск угроз до нанесения существенного урона.
<br>  Постоянный поиск угроз.
<br>  Только предпологает, где модет находиться угроза.

>**Forensic** - это криминалистика.
<br> Процесс анализа запускается уже после инцидента.
<br> Имеет входные данные об инциденте.


### Ключевые угрозы

**Удаленный доступ в инфраструктуру**

* Malware в почте
    
* Криптолокеры

* Доступ к платёжным системам

* Доступ через тестовые среды    

* Пассивные импланты -  закладки, которые доставляются и могут не проявлять себя некоторое время до тех пор пока не будут активированы. 
<br>Примером является загрузка зависимостей кода, которые разрабатывают разработчики через библиотеки или чужие коды.

* Атаки на банкоматы

**DDoS-атаки** (_Distributed Denial of Service_) 
<p style="margin-left: 10px; margin-top: -1px;">
 Ботнеры из IOT девайсов
<br> Ежегодный рост мощности атак
</p>

**Целенаправленная атака** (_advanced persistent threat_)

<img src="https://user-images.githubusercontent.com/54672403/121568745-b0e17400-ca28-11eb-8da9-14cf89b1682f.jpg" style="margin-left: 0px;height:400px">

Любая целенаправленная атака, по сути, делится на несколько этапов. 


**Первый этап** – это подготовка целенаправленной атаки. Он является обязательным и включает в себя определение цели и ее исследование.
<p style="margin-left: 10px; margin-top: 10px;">
В исследование цели может входить, исследование портала закупок, исследование сайтов организаций, которые поставляют средства защиты, для того чтобы понять, какие средства защиты установлены. <br>Может происходить подкуп или поиск сотрудников внутри организации, для того чтобы понять, как она функционирует.  
<br>В подготовительный этап также включается создание вредоносного по. Происходит использование некоторых механизмов, которые закупаются в Интернете, либо использование уязвимостей zero-day. 
<br> Далее идет проверка отсутствия срабатывания на средствах защиты
</p>


**Второй этап** – внедрение вредоносного программного обеспечения. Происходит заражение отдельных хостов. Целью его является закрепление в системе.
<p style="margin-left: 10px; margin-top: 10px;">
Заражение может происходить путем разбрасывания флешек по офису, может происходить путем доставки вредоносного программного обеспечения через почту или, например, через сайт для загрузки клиентской информации, клиентских документов для внутренних сотрудников.
</p>

**Третий этап** – распространение по инфраструктуре. 
<p style="margin-left: 10px; margin-top: 10px;">
Распространение, или lateral movement, горизонтальное распространение обычно происходит с атакованной машины. На атакованной машине получают доступ к паролю локального администратора либо взламывают его различными способами и средствами. Таким образом получается доступ на соседние машины, и осуществляется горизонтальное распространение.
</p>

**Четвертый этап** – достижение цели. 
<p style="margin-left: 10px; margin-top: 10px;">
Часто атаки происходят перед длинными выходными, когда все ушли – в пятницу вечером или в субботу утром. Время и место атаки продумывается, выбирается и заранее готовится. 
<br>В конце начинается шифрование или разрушение инфраструктуры с целью скрыть следы своего присутствия.
</p>

#### Защита периметра

- Организация межсетевого экранирования и DMZ
    - Эпизодический пересмотр правил МЭ
    - Новое правило FW
- Предотвращение/выявление сетевых атак
- Защита от DDoS-атак
- Контроль беспроводных точек доступа

#### Защита от вредоносного кода

- Антивирусная защита на уровне хостов
    - Эпизодическое выявление хостов без антивируса
    - Реагирование на заражения
- Мониторинг событий антивирусных систем
    - Антивирусная защита на уровне шлюза
    - Реагирование на заражения
- Мониторинг событий антивирусных систем


Практически у каждой организации есть внутренний почтовый сервер.
Это означает, что принятое письмо с вложением или с ссылкой будет запущено на **виртуальной инфраструктуре** песочницы.
<br>Песочница будет пытаться проанализировать поведение этого документа или файла, или будет пытаться проанализировать, что будет, если пользователь кликнет по этой ссылке, перейдет, загрузит себе вложение, и что будет делать это вложение.
<br>Фактически песочница стартует это вложение (или загруженный файл) и смотрит, как оно будет вести себя.
 
Также иногда устанавливатся **speed bump** (лежачий полицейский). Когда сотрудник или переходит по некатегорированному сайту, или сайту из определенных категорий, у него всплывает оконо. Зачастую троянское ПО не сможет перейти по этой ссылке.

#### Защита во время атаки

Стандартный антивирус может видеть отклонения от поведения прикладного ПО, может видеть обращение в таблицы LSASS или другие, SAM-таблицы, которые хранят хэши паролей администратора и может на это отреагировать
<br>Oчень важно, чтобы антивирусное ПО было правильно сконфигурировано и на него кто-то реагировал.


### MITRE matrix 

>**MITRE** -  фреймворк, который описывает этапы атак, помогает
определять на каком этапе находится атака, правильно выстраивать рубежи для каждого этапа атаки

<img src="https://user-images.githubusercontent.com/54672403/121569205-3533f700-ca29-11eb-885b-1a12bc59826d.jpg" style="margin-left: 0px;height:300px">

Первая часть включает планирование и все, что с этим связано.

<p style="margin-left: 10px; margin-top: 10px;">
Определение уязвимости организации, детектирование, подготовка к вторжению, разработка вредоносного программного обеспечения, если это требуется, разработка спирфишинга, специально сформулированного под организацию, изучение её персонала и средств защиты. 
<br>Тестирование функционала вредоносного ПО, тестирование спирфишинга на возможность преодоления защиты. 
</p>

Этапы второй части выглядят так:

|Этап | Содержание|
|----------| :------------- |
| **Initial access** | Различные способы изначального получения доступа в информационную систему.<br> Например, эксплуатация уязвимости на периметре.|
| **Persistence** | Закрепление в системе. <br> Например, создание сервисов, получение доступов к некоторым учетным записям, и другие механизмы сохранения своего состояния в этой системе. |
| **Privilege escalation** | Нужны более высокие привилегии, для того чтобы распространяться по другим информационным системам <br> Например, доступ к некоторым учетным данным.|
| **Defense evasion** | Разрушение средств защиты. |
| **Credential access** | Доступ к учетным данным. <br>Это может быть как поиск по файлам, так и эксплуатация некоторых уязвимостей, так и нелегитимный доступ в средства хранения.|
| **Discovery** | Поиск локальных файлов, изучение удаленных систем, изучение локальной системы. |
| **Lateral movement** | Горизонтальное распространение делается для того, чтобы найти целевую систему для проведения атаки или найти администраторов домена, для того чтобы было проще проводить атаку. |
| **Execution** | Например, cоздание _scheduled task_ или исполнение некоторых файлов. |
| **Collection <br>Exfiltration** | Сбор может происходить учетных записей, информации по инфраструктуре, информации по сети. Следующим этапом является уже извлечение этих данных. |
| **Command and control** |  Удаленное управление и контроль над информационной системой. |

**SIEM** - это системы включающие в себя Security Information Manager (SIM) и Security Event Manager (SEM)


<img src="https://user-images.githubusercontent.com/54672403/121569211-36652400-ca29-11eb-8348-18262d4cc7c8.jpg" style="margin-left: 0px;">

SIEM отвечает за:
- сбор логов от источников
- корреляция событий
- оповещение о событиях информационной безопасности
- генерация репортов и графиков
- долгосрочное хранение логов для расследования


**Compliance**  или управление соответствием -  отдельное направление информационной безопасности.

<p style="margin-left: 10px; margin-top: 10px;">
У кредитных организаций, довольно большое количество регуляторов. 
<br>Главным регулятором, выступает Центральный банк РФ, который выпускает положения,  требования, ГОСТы в отношении информационной безопасности
<br>Регуляции могут накладывать федеральные службы безопасности -  ФСТЭК
<br>Есть также и внешние регуляторы в виде VISA и MasterCard, которые регулируют стандарт PCI DSS.
