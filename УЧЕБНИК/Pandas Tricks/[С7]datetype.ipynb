```python
import pandas as pd
baseDF = pd.read_csv('data/dateAndString.csv', parse_dates =["date"], sep='\t')
```

## Сгенерировать дату

```python
pd.date_range(start='1/1/2018', periods=4, freq='3m')
```

## Объединить две колонки (год и день) 

```python
df = pd.DataFrame({"year": [2019, 2019, 2020],"day_of_year": [350, 365, 1]})

df["combined"] = df["year"]*1000 + df["day_of_year"]
pd.to_datetime(df["combined"], format = "%Y%j")
```

## Группировать по дате
- даты должны быть индексами

```python
df = baseDF.copy()
df.index=df.pop('date')

df.resample('Y')['sales'].sum()
# последний рабочий день месяца
df.resample('BM').apply(lambda x: x[-1])
```

# Разрезать дату

```python
df = baseDF[["date"]].copy()
date = df["date"]
df["Year"]=       date.dt.year
df["Month"]=      date.dt.month
df["Day"]=        date.dt.day
df["Hour"]=       date.dt.hour
df["Minute"]=     date.dt.minute
df["Second"]=     date.dt.second
df["Nanosecond"]= date.dt.nanosecond

df["Date"] =      date.dt.date
df["Time"] =      date.dt.time
df["Time_Zone"] = date.dt.timetz

df.head()
```

# Количество дней с начала года и получить номер дня

```python
df = baseDF[["date"]].copy()
date = df["date"]


df["Day_Of_Year"] =    date.dt.dayofyear
df["Week_Of_Year"] =   date.dt.weekofyear
df["Week"] =           date.dt.week
df["Day_Of_week"] =    date.dt.dayofweek
df["Week_Day"] =       date.dt.weekday

dayNames = {0:'Monday' , 1:'Tuesday', 2:'Wednesday',3:'Thursday' ,  4:'Friday',  5:'Saturday', 6:'Sunday'}
df["Week_Day"].replace(dayNames, inplace=True)
df.head()
```

# Дней в месяце. Високосный год. Начало месяца/года 

```python
df = baseDF[["date"]].copy()

df["Quarter"]       = df["date"].dt.quarter
df["Days_In_Month"] = df["date"].dt.days_in_month
df["Is_Month_Start"]= df["date"].dt.is_month_start
df["Is_Month_End"]     = df["date"].dt.is_month_end
df["Is_Quarter_Start"] = df["date"].dt.is_quarter_start
df["Is_Quarter_End"]   = df["date"].dt.is_quarter_end
df["Is_Leap_Year"]     = df["date"].dt.is_leap_year
df
```
