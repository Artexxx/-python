```python
import pandas as pd
df = pd.read_csv('data/dateAndString.csv', sep='\t')
df['date'] = pd.to_datetime(df.date)
```

## Форматирование таблиц по словарю

```python
fd =  {"date" : "{:%d/%m/%y}",
       "sales": "${:.2f}" }
df.head(5).style.format(fd)
```

```python
df["sales_100"] = df["sales"]*100
(
df.style.format(fd)
 .hide_index()
 .highlight_min("sales", color ="red")
 .highlight_max("sales", color ="green")
 .background_gradient(subset = "sales_100", cmap ="Blues")
 .bar("customers", color = "lightblue", align = "zero")
 .set_caption("Слишком мало стилей")
)
```
