## Построение линейных графиков
 - `x_compat=True` - делает даты наклонными
 - `subplots=True` - создаёт новый график для каждой колонки 
 - `layout=(-1, 2)` - распологает графики по 2 в ряд  

```python
df = pd.Series(np.random.randn(1000),index=ts.index).cumsum()
df.plot();
df1 = pd.DataFrame(np.random.randn(1000, 4),
                   index=ts.index,
                   columns=list('ABCD')).cumsum()
df1.plot();
df1.plot(subplots=True, layout=(-1, 2), figsize=(8, 5));
```

## Построение гистограммных графиков

```python
df2 = pd.DataFrame(np.random.randint(10, size=(15, 4)), columns=['a', 'b', 'c', 'd'])
df2.plot.bar(colormap='gist_rainbow');
df2.plot.bar(stacked=True, colormap='cubehelix');
df2.plot.barh(stacked=True, colormap='Greens');
```
